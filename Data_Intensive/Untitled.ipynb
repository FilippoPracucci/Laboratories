```python
import pandas as pd
```

```python
import os
if not os.path.exists("covid19-regioni.json"):
    from urllib.request import urlretrieve
    urlretrieve("https://raw.githubusercontent.com/pcm-dpc/COVID-19/master/dati-json/dpc-covid19-ita-regioni.json",
               "covid19-regioni.json")
```

```python
covid19_regioni = pd.read_json("covid19-regioni.json")
```

```python
covid19_regioni["denominazione_regione"]
```

```python
regioni_list = ["Emilia-Romagna", "Lombardia", "Veneto"]
my_regioni = covid19_regioni.loc[covid19_regioni["denominazione_regione"].isin(regioni_list)].sort_values("data", ascending=False)
my_regioni.data = pd.to_datetime(my_regioni.data).dt.date
my_regioni.data
```

```python
start_date = my_regioni.data.max() - pd.Timedelta(days = 5)
my_regioni = my_regioni[my_regioni.data > start_date]
my_regioni_groups = my_regioni.data.unique()
for day in my_regioni_groups:
    deceduti = my_regioni[my_regioni["data"] == day]["deceduti"].sum()
    positivi = my_regioni[my_regioni["data"] == day]["totale_positivi"].sum()
    print(deceduti / positivi * 100)
```
